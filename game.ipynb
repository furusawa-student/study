```python
class Game :
    players=[]
    chara=[]
    blue=0
    red=0
    def __init__(self,m):
        for i in range(m):
            while True :
                print("Player{}の名前の入力してください".format(i+1))
                name=input("name:")
                if name in self.players :
                    print("同じ名前は使えません")
                    continue
                else :
                    self.players.append(name)
                    break
    
    def yakusyoku(self,m):
        import random
        a=m//3
        li=random.sample(self.players,a+2)
        for i in range(a+2):
            self.chara.append(li[i])
        
    def yakusyoku2(self,l):
        while True :
            b=input("【プレイヤー確認】あなたは{}さんですか？(はいorいいえを入力してください)".format(self.players[l]))
            if b == "はい" :
                c=self.players[l]
                d=self.chara[0]
                e=self.chara[1]
                f=self.chara[2:]
                if c == d :
                    print("あなたはマーリンです")
                    print("赤陣営は以下のプレイヤーです")
                    print(self.chara[2:])
                    print("")
                
                elif c == e :
                    print("あなたはアーサー王です")
                    print("マーリンは以下のプレイヤーです")
                    print(self.chara[0])
                    print("")
                
                elif c in f :
                    print("あなたは赤陣営です")
                    print("赤陣営は以下のプレイヤーです")
                    print(self.chara[2:])
                    print("")
                
                else :
                    print("あなたは青陣営です")
                    print("")
                break
            
            elif b == "いいえ" :
                print("{}さんに渡してください".format(self.players[l]))
                continue
                
            else :
                print("はいorいいえでお答えください")
                continue

            
    def ensei(self,n) :
        print("選ばれた{}人の方は遠征の成功・失敗を決めてください".format(n))
        sof=0
        for i in range(n):
            print("{}人目の方、成功・失敗を選んでください".format(i+1))
            while True :
                a=input("「成功」または「失敗」を入力してください：")
                if a=="成功":
                    break
                elif a=="失敗":
                    sof+=1
                    break
                else :
                    continue
        if sof==0:
            self.blue+=1
            print("遠征に成功しました")
        elif sof>0:
            self.red+=1
            print("遠征に失敗しました")
```

```python
def start_game():
    inp=input("プレイヤーの人数を入力してください：")
    k=int(inp)
    game=Game(k)
    print(Game.players)
    print("上記のプレイヤーでゲームを開始します\n各プレイヤーにスマホを渡してください")
    game.yakusyoku(k)
    print(Game.yakusyoku)
    
    for i in range(k):
        game.yakusyoku2(i)
    
    print("第１回の遠征を始めます")
    game.ensei(3)
    print("")
    
    print("第２回の遠征を始めます")
    game.ensei(3)
    print("")
    
    print("第３回の遠征を始めます")
    game.ensei(4)
    print("")
    
    print("第４回の遠征を始めます")
    game.ensei(4)
    print("")
    
    print("第５回の遠征を始めます")
    game.ensei(5)
    print("")
    
    print("成功回数:{}回".format(Game.blue))
```

```python
start_game()
```
