```python
class Game :
    players=[]
    chara=[]
    blue=0
    red=0
    def __init__(self,m):
        for i in range(m):
            while True :
                print("Player{}の名前の入力してください".format(i+1))
                name=input("name:")
                if name in self.players :
                    print("同じ名前は使えません")
                    continue
                else :
                    self.players.append(name)
                    break
    
    def yakusyoku(self,m):
        import random
        a=m//3
        li=random.sample(self.players,a+2)
        for i in range(a+2):
            self.chara.append(li[i])
        
    def yakusyoku2(self,l):
        while True :
            b=input("【プレイヤー確認】あなたは{}さんですか？(はいorいいえを入力してください)".format(self.players[l]))
            if b == "はい" :
                c=self.players[l]
                d=self.chara[0]
                e=self.chara[1]
                f=self.chara[2:]
                if c == d :
                    print("メモ帳を確認してください")
                    file = open('InputText.txt', 'w')
                    file.write("あなたはマーリンです。赤陣営は以下のプレイヤーです。\n")
                    file.write(str(self.chara[2:]))
                    file.write("\nメモ帳を閉じてください")
                    subprocess.Popen([r'C:\WINDOWS\system32\notepad.exe',IT])
                    #file.close()
                    #print("あなたはマーリンです")
                    #print("赤陣営は以下のプレイヤーです")
                    #print(self.chara[2:])
                    #print("")
                
                elif c == e :
                    print("メモ帳を確認してください")
                    file = open("InputText.txt","w")
                    file.write("あなたはアーサー王です。マーリンは以下のプレイヤーです。\n")
                    file.write(str(self.chara[0]))
                    file.write("\nメモ帳を閉じてください")
                    subprocess.Popen([r'C:\WINDOWS\system32\notepad.exe',IT])
                    #print("あなたはアーサー王です")
                    #print("マーリンは以下のプレイヤーです")
                    #print(self.chara[0])
                    #print("")
                
                elif c in f :
                    print("メモ帳を確認してください")
                    file=open("InputText.txt","w")
                    file.write("あなたは赤陣営です。赤陣営は以下のプレイヤーです。\n")
                    file.write(str(self.chara[2:]))
                    file.write("\nメモ帳を閉じてください")
                    subprocess.Popen([r'C:\WINDOWS\system32\notepad.exe',IT])
                    #print("あなたは赤陣営です")
                    #print("赤陣営は以下のプレイヤーです")
                    #print(self.chara[2:])
                    #print("")
                
                else :
                    print("メモ帳を確認してください")
                    file=open("InputText.txt","w")
                    file.write("あなたは青陣営です。")
                    file.write("メモ帳を閉じてください")
                    subprocess.Popen([r'C:\WINDOWS\system32\notepad.exe',IT])
                    #print("あなたは青陣営です")
                    #print("")
                break
            
            elif b == "いいえ" :
                print("{}さんに渡してください".format(self.players[l]))
                continue
                
            else :
                print("はいorいいえでお答えください")
                continue

            
    def ensei(self,n) :
        if self.blue <= 2 and self.red <= 2 :
            print("選ばれた{}人の方は遠征の成功・失敗を決めてください".format(n))
            sof=0
            for i in range(n):
                print("{}人目の方、成功・失敗を選んでください".format(i+1))
                while True :
                    a=getpass("「s」または「f」を入力してください\n「s」を入力すると遠征は成功、「f」を入力すると失敗します：")
                    if a=="s":
                        break
                    elif a=="f":
                        sof+=1
                        break
                    else :
                        print("「s」か「f」を入力してください")
                        continue
            if sof==0:
                self.blue+=1
                print("遠征に成功しました")
                print("青陣営の遠征成功回数は{}回です。".format(self.blue))
                if self.blue>=3 :
                    print("青陣営が大遠征に成功しました")
                    
                    print("赤陣営は誰か一人を暗殺できます。\n暗殺する人を選んでください\nマーリンを暗殺した場合、青陣営の大遠征の成功は無効となり、赤陣営の勝利となります")
                    a=input("暗殺する人を選んでください：")
                    if a==self.chara[0] :
                        print("マーリンが暗殺されました。\n赤陣営の勝利です")
                    else :
                        print("マーリンは暗殺されませんでした。\n青陣営の勝利です")
                    
            elif sof>0:
                self.red+=1
                print("遠征に失敗しました")
                print("赤陣営の遠征阻止回数は{}回です。".format(self.red))
                if self.red>=3 :
                    print("赤陣営が大遠征を阻止しました")
        else :
            print("片方の陣営が勝利条件を満たしたため、この遠征はスキップされます")
            
    def ensei2(self,n) :
        if self.blue <= 2 and self.red <= 2 :
            print("選ばれた{}人の方は遠征の成功・失敗を決めてください".format(n))
            sof=0
            for i in range(n):
                print("{}人目の方、成功・失敗を選んでください".format(i+1))
                while True :
                    a=getpass("「s」または「f」を入力してください\n「s」を入力すると遠征は成功、「f」を入力すると失敗します：")
                    if a=="s":
                        break
                    elif a=="f":
                        sof+=1
                        break
                    else :
                        print("「s」か「f」を入力してください")
                        continue
            if sof<=1:
                self.blue+=1
                print("遠征に成功しました")
                print("青陣営の遠征成功回数は{}回です。".format(self.blue))
                if self.blue>=3 :
                    print("青陣営が大遠征に成功しました")
                    
                    print("赤陣営は誰か一人を暗殺できます。\n暗殺する人を選んでください\nマーリンを暗殺した場合、青陣営の大遠征の成功は無効となり、赤陣営の勝利となります")
                    a=input("暗殺する人を選んでください：")
                    if a==self.chara[0] :
                        print("マーリンが暗殺されました。\n赤陣営の勝利です")
                    else :
                        print("マーリンは暗殺されませんでした。\n青陣営の勝利です")
            elif sof>1:
                self.red+=1
                print("遠征に失敗しました")
                print("赤陣営の遠征阻止回数は{}回です。".format(self.red))
                if self.red>=3 :
                    print("赤陣営が大遠征を阻止しました")
        else :
            print("片方の陣営が勝利条件を満たしたため、この遠征はスキップされます")
```

```python
def start_game():
    inp=input("プレイヤーの人数を入力してください：")
    k=int(inp)
    game=Game(k)
    print(Game.players)
    print("上記のプレイヤーでゲームを開始します\n各プレイヤーにスマホを渡してください")
    game.yakusyoku(k)
    print(Game.yakusyoku)
    
    for i in range(k):
        game.yakusyoku2(i)
    
    print("第１回の遠征を始めます")
    game.ensei(3)
    print("")
    
    print("第２回の遠征を始めます")
    game.ensei(3)
    print("")
    
    print("第３回の遠征を始めます")
    game.ensei(4)
    print("")
    
    print("第４回の遠征を始めます")
    game.ensei(4)
    print("")
    
    print("第５回の遠征を始めます")
    game.ensei2(5)
    print("")
    
    #print("成功回数:{}回".format(Game.blue))
```
